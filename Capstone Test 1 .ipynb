```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns




df = pd.read_csv('data_ohe.csv')
```

```python
df['Status'].value_counts()
```

In future iterations of this notebook we will be looking into feature engineering as such for now we will drop the collumns bellow that we will not be able to apply into our regression models. 

```python
df=df.drop(['Listing Pictures','Unnamed: 0','ML #','City','Postal Code','S/A',\
              'Address','Depth','TypeDwel','Style of Home','Fixtures Removed Y/N',\
              'Fixtures Remvd','Last Trans Date','Distance to Pub/Rapid Tr',\
              'Dist to School/School Bus','Sold Price','Sold Date','Type',\
              'List_Price_num','Parking Places - Total','List Date'],axis = 1)
```

```python
#Sanity Check 
df.select_dtypes(include='object').columns
```

```python
df.isna().sum()
```

## Linear Regression 

```python
X = df.drop(['Price_num'], axis=1)
y = df['Price_num'] 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size = 0.25,
                     random_state=1)
```

```python
import statsmodels.api as sm
X_withconstant = sm.add_constant(X_train) #we have to add in our intercept manually!

```

```python
# 1. Instantiate Model
myregression = sm.OLS(y_train, X_withconstant)

# 2. Fit Model (this returns a seperate object with the parameters)
myregression_results = myregression.fit()

# Looking at the summary
myregression_results.summary()

# ignore red warning below
```

## KNN regressor 

```python
from sklearn.neighbors import KNeighborsRegressor
KNNmodel = KNeighborsRegressor(weights='distance')
KNNmodel.fit(X_train, y_train)
```

```python
print(f"R^2 score on training set: {KNNmodel.score(X_train, y_train)}")
print(f"R^2 score on testing set: {KNNmodel.score(X_test, y_test)}")
```

## Gradient Boosting 

```python
from sklearn import ensemble
GBR = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 20, min_samples_split = 2,
          learning_rate = 0.1, loss = 'ls')
```

```python
GBR.fit(X_train, y_train)
```

```python
print(f"R^2 score on training set: {GBR.score(X_train,y_train)}")
print(f"R^2 score on testing set: {GBR.score(X_test,y_test)}")
```

## XG Boosting 

```python
import xgboost as xg
```

```python
XGB = xg.XGBRegressor()

XGB.fit(X_train, y_train)
print(XGB.score(X_train, y_train))
print(XGB.score(X_test,y_test))
```

These are the baseline of all our models. We see that they are all performing relatively well. However will need to be tuned significantly in order to find the best performing model. Once we quickly hyperparametize we will focus on one model to move forward. So far the best performing model is the XG boosting model.
